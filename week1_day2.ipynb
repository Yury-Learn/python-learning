````python
# Эта ячейка:
# 1. Спрашивает число у пользователя.
# 2. Превращает введённую строку в целое число (int).
# 3. Проверяет чётность через оператор % (остаток от деления).

# raw_number: строка, которую пользователь ввёл.
# input("Введи число: ") показывает приглашение "Введи число: "
# и ждёт пока пользователь введёт что-то и нажмёт Enter.
raw_number = input("Введи число: ")

# int(raw_number) пытается преобразовать строку в целое число.
# Пример:
#   "10" -> 10
#   "42" -> 42
# Если пользователь введёт, например, "кот", будет ошибка ValueError,
# потому что строку "кот" нельзя преобразовать к числу.
number = int(raw_number)

# number % 2 — это остаток от деления на 2.
# Примеры:
#   10 % 2 == 0  -> число чётное
#   11 % 2 == 1  -> число нечётное
if number % 2 == 0:
    print("Число чётное")
else:
    print("Число нечётное")

# Для наглядности выведем типы:
print("Сырые данные (raw_number):", raw_number, "| тип:", type(raw_number))
print("После преобразования (number):", number, "| тип:", type(number))
````

## Week1 Day3 — if, пароли, оператор сравнения

### 1. Проверка пароля через `if`

Логика:
1. Я задаю ожидаемый пароль в переменную `correct_password`.
2. Пользователь вводит строку через `input()`.
3. Мы сравниваем введённое значение с ожидаемым.

Пример:

```python
correct_password = "qwerty"
user_input = input("Введи пароль: ")

if user_input == correct_password:
    print("Доступ разрешён")
else:
    print("Доступ запрещён")
    
Разбор:

if <условие>: — проверить условие.

else: — что делать, если условие НЕ выполнилось.

user_input == correct_password — сравнение двух строк.

2. Разница между = и ==
= (один знак равенства) — ПРИСВОЕНИЕ значения переменной.

Пример: x = 10 означает «положить 10 в переменную x».

== (два знака равенства) — СРАВНЕНИЕ.

Пример: x == 10 означает «проверить, действительно ли x равно 10».

Результат сравнения — это булево значение True или False (тип bool).

3. Почему хранить пароль прямо в коде — плохо в реальной инфраструктуре
correct_password = "qwerty" — это пример ДЛЯ ОБУЧЕНИЯ.

В живой инфраструктуре так делать нельзя, потому что:

Любой, у кого есть доступ к коду или к истории Git, увидит пароль.

Если код случайно попадёт в публичный GitHub — пароль утечёт.

Пароль нельзя централизованно поменять (он вшит в код).

Как делают правильно в проде:

Не хранят пароль в открытом виде, хранят только хэш (цифровой отпечаток).

Сравнивают не пароли, а хэши.

Либо берут секреты из переменных окружения или менеджера секретов (а не прописывают строкой в коде).

То, что мы делаем здесь, — это учебный пример, чтобы понять if, == и ветвления.

````python
# Демонстрация разницы между присвоением (=) и сравнением (==)

x = 10  # присвоили значение 10 переменной x
print("x сейчас =", x)

check_equal_10 = (x == 10)   # True, потому что x это 10
check_equal_99 = (x == 99)   # False, потому что x не 99

print("x == 10 даёт:", check_equal_10, "| тип:", type(check_equal_10))
print("x == 99 даёт:", check_equal_99, "| тип:", type(check_equal_99))

# Важно:
# - Оператор = меняет переменную (присваивает).
# - Оператор == ничего не меняет, он просто сравнивает и даёт True/False.
````
